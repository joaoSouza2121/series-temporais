```r
install.packages("forecast")
```

```r
library(forecast)
```

Analisando A série dados temporais Airpassenger - classica para estudos

Conhecendo os Dados

```r
AirPassengers
start(AirPassengers)
end(AirPassengers)
frequency(AirPassengers)
```

Plots

```r
plot(AirPassengers)
plot(aggregate(AirPassengers))
```

Gerando uma janela para analisar

```r
subst <- window(AirPassengers, start = c(1960,1), end = c(1960,12))
subst
plot(subst)
```

Decomposição da Serie

```r
dec = decompose(AirPassengers)
dec 
```

Atributos

```r
attributes(dec)
```

Efeito sazonal

```r
dec$seasonal
```

Analisando os Elementos da Decomposição

```r
plot(dec)
```

Tendencia

```r
plot(dec$trend)
```

Elemento Aleatório - Random

```r
plot(dec$random)
```

A Serie Original

```r
plot(dec$x)
```

Suavização Exponencial

```r
ets <- ets(AirPassengers)
ets 
```

Gerando Previsão

```r
previsao <- forecast(ets, h = 12)
plot(previsao)
```

Previsão com Arima

```r
arima <- auto.arima(AirPassengers)
```

```r
previsao_arima <- forecast(arima, h = 12)
plot(previsao_arima)
```
